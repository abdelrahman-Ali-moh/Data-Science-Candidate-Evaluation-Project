# EDA Class

```python
class MySQLDataAnalyzer:
    def __init__(self, host, user, password, database):
        self.db_config = {
            'host': host,
            'user': user,
            'password': password,
            'database': database
        }
        self.conn = None

    def connect_to_mysql(self):
        self.conn = mysql.connector.connect(**self.db_config)

    def close_mysql_connection(self):
        if self.conn:
            self.conn.close()

    def perform_eda(self, mysql_table):
        try:
            # Connect to MySQL
            self.connect_to_mysql()

            # Fetch data into a Pandas DataFrame
            sql_query = f"SELECT * FROM {mysql_table}"
            df = pd.read_sql_query(sql_query, self.conn)

            # Display the first few rows of the dataset
            #print(df.head())

            # Get general information about the dataset
            #print(df.info())

            # Summary statistics
            df.describe()

            # Visualize missing values
            df.isnull().sum()

            # Drop missing values and modify the original DataFrame
            df = df.dropna()

            return df
        except Exception as e:
            print(f"Error: {e}")
        finally:
            # Close MySQL connection
            self.close_mysql_connection()

```

# Customer table EDA

```python
class DataVisualizer_Customer:
    def __init__(self, df):
        self.df = df

    def visualize_customer_state_counts(self):
        total = self.df['customer_state'].count()
        vc = self.df['customer_state'].value_counts()

        ax = sns.countplot(data=self.df, x='customer_state')
        plt.xticks(rotation=45)
        for p in ax.patches:
            ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + 0.6, p.get_height() + 80),
                        ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=90)

    def visualize_top_customer_cities(self):
        top_cities = self.df['customer_city'].value_counts().nlargest(5)
        plt.figure(figsize=(10, 6))
        sns.barplot(x=top_cities.index, y=top_cities.values)
        plt.title('Top 5 Customer Cities')
        plt.xlabel('Customer City')
        plt.ylabel('Count')

    def show_plots(self):
        plt.show()
```

```python
analyzer = MySQLDataAnalyzer(host='localhost', user='root', password='2556250', database='od')
df = analyzer.perform_eda(mysql_table='customers')
visualizer = DataVisualizer_Customer(df)
visualizer.visualize_customer_state_counts()
visualizer.visualize_top_customer_cities()
visualizer.show_plots()
```

# Geo table EDA

```python
class GeolocationVisualizer:
    def __init__(self, df):
        self.df = df

    def visualize_latitude_longitude_distribution(self):
        plt.figure(figsize=(12, 6))

        # Distribution of Latitude
        plt.subplot(1, 2, 1)
        sns.histplot(self.df['geolocation_lat'], bins=20, kde=True)
        plt.title('Distribution of Latitude')

        # Distribution of Longitude
        plt.subplot(1, 2, 2)
        sns.histplot(self.df['geolocation_lng'], bins=20, kde=True)
        plt.title('Distribution of Longitude')

        plt.tight_layout()
        plt.show()

    def explore_relationships(self):
        # Pairplot for exploring relationships between variables
        sns.pairplot(self.df[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']])
        plt.show()

    def countplot_states(self):
        # Countplot for states
        plt.figure(figsize=(8, 5))
        sns.countplot(x='geolocation_state', data=self.df)
        plt.title('Count of Records by State')
        plt.show()
```

```python
analyzer = MySQLDataAnalyzer(host='localhost', user='root', password='2556250', database='od')
df = analyzer.perform_eda(mysql_table='geolocation')
geolocation_visualizer = GeolocationVisualizer(df)
geolocation_visualizer.visualize_latitude_longitude_distribution()
geolocation_visualizer.explore_relationships()
geolocation_visualizer.countplot_states()
```

# Orders

```python
class OrderVisualizer:
    def __init__(self, df):
        self.df = df

    def visualize_order_status_distribution(self):
        # Visualize the distribution of order_status using a pie chart
        order_status_counts = self.df['order_status'].value_counts()
        labels = order_status_counts.index
        sizes = order_status_counts.values

        # Create a pie chart
        fig, ax = plt.subplots(figsize=(10, 8))
        pie = ax.pie(sizes, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

        # Create a table outside the pie chart
        table_data = [['Order Status', 'Count']]
        table_data.extend(zip(labels, sizes))
        table = plt.table(cellText=table_data, loc='bottom', colLabels=None, cellLoc='center', bbox=[0, -0.4, 1, 0.35])

        # Adjust the appearance of the table
        table.auto_set_font_size(False)
        table.set_fontsize(10)

        # Title for the entire plot
        plt.title('Distribution of Order Status')
        plt.show()

    def visualize_order_purchase_timestamp_distribution(self):
        # Visualize the distribution of order_purchase_timestamp
        self.df['order_purchase_timestamp'] = pd.to_datetime(self.df['order_purchase_timestamp'])
        plt.figure(figsize=(14, 6))
        sns.histplot(self.df['order_purchase_timestamp'], bins=50, kde=True)
        plt.title('Distribution of Order Purchase Timestamp')
        plt.xlabel('Order Purchase Timestamp')
        plt.ylabel('Count')
        plt.show()
```

```python
analyzer = MySQLDataAnalyzer(host='localhost', user='root', password='2556250', database='od')
df = analyzer.perform_eda(mysql_table='orders')
order_visualizer = OrderVisualizer(df)
order_visualizer.visualize_order_status_distribution()
order_visualizer.visualize_order_purchase_timestamp_distribution()
```

## Product table Prints unique product categories.
### Prints the counts of each product category.
### Visualizes the distribution of product categories using a countplot.
### Prints summary statistics (mean, std, min, max, etc.) for numerical columns grouped by product category.
### Creates a boxplot to visualize the distribution of product name lengths by category.

```python
class ProductCategoryAnalyzer:
    def __init__(self, df):
        self.df = df

    def check_unique_categories(self):
        # Check the unique values in the 'product_category_name' column
        unique_categories = self.df['product_category_name'].unique()
        print("Unique Categories:", unique_categories)

    def count_and_visualize_categories(self):
        # Count the occurrences of each product category
        category_counts = self.df['product_category_name'].value_counts()
        print("Category Counts:\n", category_counts)

        # Visualize the distribution of product categories
        plt.figure(figsize=(12, 6))
        sns.countplot(x='product_category_name', data=self.df, order=category_counts.index)
        plt.xticks(rotation=90)
        plt.title('Distribution of Product Categories')
        plt.xlabel('Product Category')
        plt.ylabel('Count')
        plt.show()

    def summarize_numerical_by_category(self):
        # Summary statistics for numerical columns grouped by product category
        category_summary = self.df.groupby('product_category_name').describe().transpose()
        print("Summary Statistics by Product Category:\n", category_summary)
```

```python
analyzer = MySQLDataAnalyzer(host='localhost', user='root', password='2556250', database='od')
df = analyzer.perform_eda(mysql_table='products')
product_category_analyzer = ProductCategoryAnalyzer(df)
product_category_analyzer.check_unique_categories()
product_category_analyzer.count_and_visualize_categories()
product_category_analyzer.summarize_numerical_by_category()
```

## Displays basic information about reviews table.
** Shows summary statistics for numerical columns.
*Visualizes the distribution of review scores.
*Displays the most common review scores.
*Visualizes the distribution of review scores over time.
*Generates a word cloud for review comments.

```python
class ReviewAnalyzer:
    def __init__(self, df):
        self.df = df

    def visualize_review_scores_distribution(self):
        # Visualize the distribution of review scores
        sns.countplot(x='review_score', data=self.df)
        plt.title('Distribution of Review Scores')
        plt.show()

    def display_top_review_scores(self):
        # Display the most common review scores
        top_review_scores = self.df['review_score'].value_counts().head(5)
        print("Top Review Scores:")
        print(top_review_scores)

    def visualize_review_scores_over_time(self):
        # Visualize the distribution of review scores over time
        self.df['review_creation_date'] = pd.to_datetime(self.df['review_creation_date'])
        self.df['year_month'] = self.df['review_creation_date'].dt.to_period('M')
        sns.countplot(x='year_month', hue='review_score', data=self.df)
        plt.title('Distribution of Review Scores Over Time')
        plt.xticks(rotation=45)
        plt.show()

    def display_word_cloud_for_comments(self):
        # Display word cloud for review comments
        comment_text = ' '.join(self.df['review_comment_message'].dropna())
        wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(comment_text)

        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud for Review Comments')
        plt.show()
```

```python
analyzer = MySQLDataAnalyzer(host='localhost', user='root', password='2556250', database='od')
df = analyzer.perform_eda(mysql_table='reviews')
review_analyzer = ReviewAnalyzer(df)
review_analyzer.visualize_review_scores_distribution()
review_analyzer.display_top_review_scores()
review_analyzer.visualize_review_scores_over_time()
review_analyzer.display_word_cloud_for_comments()
```
